```python
import pandas as pd
import re

movies_df = pd.read_csv('movies_stats.csv', dtype_backend='numpy_nullable', usecols=['rating', 'title'])
movies_df.info()
```

```python
movies_df.head()
```

```python
years = list(range(1950, 2011))
def production_year(title):
    """Get release year by movie's title."""
    year = 1900
    match = re.search(r'\((\d{4})\)$', title)
    if match:
        year_match = int(match.group(1))
        year = year_match if year_match in years else year
    return year
```

```python
movies_df['year'] = movies_df['title'].map(production_year)
movies_df.head()
```

```python
movies_df_avg = movies_df.groupby('year').agg(avgyear_rating=pd.NamedAgg('rating', 'mean'))
movies_df_sorted = movies_df_avg.sort_values(by='avgyear_rating', ascending=False)
movies_df_sorted.head(20)
```
